## Subsets
Question: 
>  <br>

Example:

Key Idea:
> To generate all possible subsets of a given set of distinct integers, we can use a recursive approach. <br>
> For each element, we have two choices: either include it in the current subset or exclude it. <br>
> We explore both choices recursively to generate all subsets. <br>

```python
from typing import List

class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        def backtrack(start: int, subset: List[int]):
            # Add the current subset to the result
            subsets.append(subset[:])  # Append a copy of the subset
            
            # Explore further by including additional elements
            for i in range(start, len(nums)):
                subset.append(nums[i])      # Include nums[i] in the subset
                print(f"start:{nums[i]}, subset: {subset}, subsets: {subsets}")
                backtrack(i + 1, subset)    # Recursively call for the next elements
                subset.pop()                # Backtrack: remove nums[i] from the subset
        
        subsets = []
        backtrack(0, [])
        return subsets

# Example usage
solution = Solution()
nums = [1, 2, 3]
result = solution.subsets(nums)
print(result)
```

## Letter Combinations of a Phone Number
Question: 
>  <br>

Example:

Key Idea:
> 

```python
from typing import List

class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        # If input is empty, return an empty list
        if not digits:
            return []

        # Mapping of digits to corresponding letters
        phone_map = {
            "2": "abc", "3": "def", "4": "ghi", "5": "jkl",
            "6": "mno", "7": "pqrs", "8": "tuv", "9": "wxyz"
        }

        # Result to store all combinations
        res = []

        # Helper function for backtracking
        def backtrack(index, path):
            # If the current path length equals the number of digits, add to result
            if index == len(digits):
                res.append("".join(path))
                return
            
            # Get letters corresponding to the current digit
            possible_letters = phone_map[digits[index]]
            for letter in possible_letters:
                # Add letter to the path
                path.append(letter)
                print(f"start:{letter}, subset: {path}, subsets: {res}")
                # Recurse to the next digit
                backtrack(index + 1, path)
                # Backtrack by removing the last letter
                path.pop()

        # Start backtracking from the first digit
        backtrack(0, [])
        return res

# Example usage
solution = Solution()
digits = "23"
result = solution.letterCombinations(digits)
print(result)
```
