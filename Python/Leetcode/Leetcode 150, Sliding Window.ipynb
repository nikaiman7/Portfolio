## Best Time to Buy and Sell Stock
Question: 
>  <br>

Example:
>  <br>

Key Idea:
> To find the maximum profit from a single buy and sell transaction in the input array 'prices', we can use a simple one-pass approach. <br>
> We initialize two variables, 'min_price' to keep track of the minimum price seen so far, and 'max_profit' to store the maximum profit.<br>
> We iterate through the 'prices' array, and for each price, we update the 'min_price' if we find a smaller price.<br>
> We also calculate the potential profit if we sell at the current price and update 'max_profit' if the current profit is greater than the previous maximum profit.<br> 

```python
from typing import List

class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        if not prices:
            return 0

        min_price = float("inf")
        max_profit = 0

        for price in prices:
            min_price = min(min_price, price)
            max_profit = max(max_profit, price - min_price)
            print(f"price: {price}, min_price: {min_price}, max_profit:{max_profit}")

        return max_profit

solution = Solution()

prices = [7,1,5,3,6,4]
print(solution.maxProfit(prices)) 
```

## Longest Substring Without Repeating Characters
Question: 
> Given a string s, find the length of the longest substring without repeating characters. <br>

Example:
>  <br>

Key Idea:
> To find the length of the longest substring without repeating characters in the input string 's', we can use the sliding window approach.<br>
> We use two pointers, 'left' and 'right', to represent the current window.<br>
> As we move the 'right' pointer to the right, we expand the window and add characters to a set to keep track of unique characters in the window.<br>
> If we encounter a repeating character, we move the 'left' pointer to the right to shrink the window until the repeating character is no longer in the window.<br>
> At each step, we update the maximum length of the window (i.e., the length of the longest substring without repeating characters). <br>

```python
def lengthOfLongestSubstring(s: str) -> int:
    left, right = 0, 0
    max_length = 0
    unique_chars = set()
    print(f"str: {s}")

    while right < len(s):
        if s[right] not in unique_chars:
            unique_chars.add(s[right])
            max_length = max(max_length, right - left + 1)
            right += 1
        else:
            unique_chars.remove(s[left])
            left += 1
        print(f"left: {left}, right: {right}, max_length:{max_length}, unique_chars: {unique_chars}")

    return max_length

s = "abcabcbb"
print(lengthOfLongestSubstring(s))

s = "bbbbb"
print(lengthOfLongestSubstring(s))

s = "pwwkew"
print(lengthOfLongestSubstring(s))
```

## Longest Repeating Character Replacement
Question: 
>  <br>

Example:
>  <br>

Key Idea:
>  <br>

```python
def characterReplacement(s: str, k: int) -> int:
    left, right = 0, 0
    max_length = 0
    char_freq = {}
    max_freq = 0

    while right < len(s):
        char_freq[s[right]] = char_freq.get(s[right], 0) + 1
        max_freq = max(max_freq, char_freq[s[right]])

        if (right - left + 1) - max_freq > k:
            char_freq[s[left]] -= 1
            left += 1

        max_length = max(max_length, right - left + 1)
        right += 1
        print(f"left: {left}, right: {right}, window: {(right - left + 1) - max_freq}, max_length:{max_length}, max_freq: {max_freq}, char_freq: {char_freq}")

    return max_length

s, k = "ABAB", 2
print(characterReplacement(s, k))

s, k = "AABABBA", 1
print(characterReplacement(s, k))   
```
