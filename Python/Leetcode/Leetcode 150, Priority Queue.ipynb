## Kth Largest Element in a Stream
Question: 
>  <br>

Example:
> 

Key Idea:
>  <br>

```python
import heapq

class KthLargest:
    def __init__(self, k: int, nums: list[int]):
        self.k = k
        self.heap = nums
        heapq.heapify(self.heap)        # Convert nums into a min-heap
        while len(self.heap) > k:
            heapq.heappop(self.heap)    # Maintain heap size of k
    
    def add(self, val: int) -> int:
        heapq.heappush(self.heap, val)  # Add the new value
        if len(self.heap) > self.k:
            heapq.heappop(self.heap)    # Maintain heap size of k
        return self.heap[0]             # The root of the heap is the kth largest

# Example Usage
if __name__ == "__main__":
    # Test Input
    operations = ["KthLargest", "add", "add", "add", "add", "add"]
    arguments = [[3, [4, 5, 8, 2]], [3], [5], [10], [9], [4]]
    
    # Initialize KthLargest object
    kthLargest = KthLargest(arguments[0][0], arguments[0][1])
    results = [None]  # The result of initializing the object is `null`

    # Perform `add` operations
    for i in range(1, len(operations)):
        result = kthLargest.add(arguments[i][0])
        results.append(result)
    
    print(results)  # Expected Output: [null, 4, 5, 5, 8, 8]
```

## Last Stone Weight
Question: 
>  <br>

Example:
> 

Key Idea:
>  <br>

```python
import heapq

class Solution:
    def lastStoneWeight(self, stones: list[int]) -> int:
        # Convert stones to a max-heap (by pushing negative values)
        max_heap = [-stone for stone in stones]
        heapq.heapify(max_heap)
        print(f"start max_heap: {max_heap}")
        
        # Continue until there is at most one stone left
        while len(max_heap) > 1:
            # Pop the two heaviest stones
            stone1 = -heapq.heappop(max_heap)  # Heaviest stone
            stone2 = -heapq.heappop(max_heap)  # Second heaviest stone
            
            # If they are not equal, push the remaining weight back into the heap
            if stone1 != stone2:
                heapq.heappush(max_heap, -(stone1 - stone2))

            print(f"Heaviest stone: {stone1}, Second heaviest stone: {stone2}, Delta: {(stone1 - stone2)}, current max_heap: {max_heap}")
        
        # If there is one stone left, return its weight, otherwise return 0
        return -max_heap[0] if max_heap else 0
    
solution = Solution()
stones = [2, 7, 4, 1, 8, 1]
solution.lastStoneWeight(stones)
```
