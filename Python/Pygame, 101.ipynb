```python
# Import Libraries
import pygame 
from pygame import mixer

# Initiate Pygame module
pygame.init() 

# Color & position
white = (255,255,255) 
red   = (255,0,0)   
blue  = (0,0,255)    
black = (0,0,0)   
green = (0,255,0)
position = (100,50) 

# Create a window of desired size (width, height)
# canvas = pygame.display.set_mode((500, 500))                     # Not resizable
canvas = pygame.display.set_mode((500, 500), pygame.RESIZABLE)   # Resizable
  
# Set the title of the window/ board
pygame.display.set_caption("Show Image") 

# image
image  = pygame.image.load("../Python/Input/GeeksforGeeks.png") 
# image2 = pygame.transform.rotate(image, 180)    # rotate image
image2 = pygame.transform.flip(image, True, False)    # flip image // pygame.transform.flip(Surface, xbool, ybool)

exit = False  
while not exit: 
    # Set canvas color
    canvas.fill(black)                           # Fill with selected color

    # Blitting
    canvas.blit(image, dest = position)          # Render image on to screen
    pygame.Surface.set_colorkey(image, white)    # Make the whie part transparent
    canvas.blit(image2, (position[0], position[1]+500))        
    image2.convert()                              # change the pixel format of an image
    canvas.blit(image2, (position[0]+500, position[1])) 

    # Draw
    # solid rectangle
    pygame.draw.rect(canvas, red,         
                     pygame.Rect(position[0]+100,position[1]+30,60,60), 0) 
    # outlined rectangle
    pygame.draw.rect(canvas, red,         
                     pygame.Rect(position[0]+350,position[1]+30,60,60), 3) 
    # outlined rectangle with rounded corner
    pygame.draw.rect(canvas, red,         
                     pygame.Rect(position[0]+250,position[1]+30,60,60), 3, 3) 
    # solid circle
    pygame.draw.circle(canvas, blue,                
                       (position[0]+100, position[1]+200), 10, 0) 
    # outlined circle 
    pygame.draw.circle(canvas, blue,               
                       (position[0]+350, position[1]+200), 10, 3)
    # solid polygon
    pygame.draw.polygon(canvas, green,             
                        [[position[0]+300, position[1]+300], [position[0]+100, position[1]+400],
                         [position[0]+100, position[1]+300]], 0)
    # outlined polygon
    pygame.draw.polygon(canvas, green,             
                    [[position[0]+300, position[1]+300], [position[0]+450, position[1]+400],
                     [position[0]+450, position[1]+300]], 3)
    # line
    pygame.draw.line(canvas, white,                
                     [position[0]+100, position[1]+150], 
                     [position[0]+450, position[1]+150], 5)

    # Draw shapes with mouse
    circle_pos = []                               # store position of circles
    circle_rad = 60                               # radius of circle
    
    # # Clock
    # clock = pygame.time.Clock()                  # Create clock obj
    # print(clock.get_time())                      # printing time used in the previous tick
    # print(clock.get_fps())                       # printing compute the clock framerate                        

    # Events
    for event in pygame.event.get():              # To empty the event queue. If we do not call this, the game will become unresponsive.
        # Event termination
        if event.type == pygame.QUIT:             
            exit = True
        
        # Mouse poisitioning
        if event.type == pygame.MOUSEBUTTONDOWN:
            position = pygame.mouse.get_pos()

        # Keyboard
        if event.type == pygame.KEYDOWN:
            # Experiment with screen color & text
            if event.key == pygame.K_UP:
                canvas.fill(blue)                            
                
                font = pygame.font.SysFont('../Python/Input/FreeSansBold.ttf', 100)
                text = font.render('YOU PRESSED UP', True, red)
                textRect = text.get_rect()
                textRect.center = (300, 200)
                canvas.blit(text,textRect)

                pygame.display.flip()
                pygame.time.wait(1000)                       
            if event.key == pygame.K_DOWN:
                canvas.fill(red)    
                
                font = pygame.font.SysFont('../Python/Input/FreeSansBold.ttf', 100)
                text = font.render('YOU PRESSED DOWN', True, blue)
                textRect = text.get_rect()
                textRect.center = (300, 200)
                canvas.blit(text,textRect)

                pygame.display.flip()
                pygame.time.wait(1000)                     

            # Experiment with audio
            if event.key == pygame.K_1:
                mixer.init() 
                mixer.music.load("../Python/Input/pirates.mp3") 
                mixer.music.set_volume(0.7) 
                mixer.music.play()
            if event.key == pygame.K_2:
                mixer.music.pause()
            if event.key == pygame.K_3:
                mixer.music.unpause()
            if event.key == pygame.K_4:
                mixer.music.stop()
                break
    
    # Update the GUI game
    pygame.display.update() 

# Quit pygame
pygame.display.quit()                                 
```

## Text Writing

```python
# import pygame module
import pygame
 
# import time module
import time
 
# initialize the pygame module
pygame.init()
 
# set the window screen size
display_screen = pygame.display.set_mode((500, 500))
 
# add some text
text = 'Type Something: '
 
# add default font style with font
# size
font = pygame.font.SysFont(None, 40)
 
# render the text
img = font.render(text, True, (255, 0, 0))
 
rect = img.get_rect()
rect.topleft = (20, 20)
cursor = pygame.Rect(rect.topright, (3, rect.height))
 
running = True
 
while running:
     
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
 
        # detect if key is physically 
        # pressed down
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_BACKSPACE:
                if len(text) > 0:
                     
                    # stores the text except last
                    # character
                    text = text[:-1]
 
            else:
                text += event.unicode
                 
            img = font.render(text, True, (255, 0, 0))
            rect.size = img.get_size()
            cursor.topleft = rect.topright
 
    # Add background color to the window screen
    display_screen.fill((200, 255, 200))
    display_screen.blit(img, rect)
     
    # cursor is made to blink after every 0.5 sec
    if time.time() % 1 > 0.5:
        pygame.draw.rect(display_screen, (255, 0, 0), cursor)
         
    # update display
    pygame.display.update()
 
pygame.quit()
```

## Moving Image

```python
# Import the library pygame
import pygame
from pygame.locals import *

# Take colors input
YELLOW = (255, 255, 0)
BLUE = (0, 0, 255)

# Construct the GUI game
pygame.init()

# Set dimensions of game GUI
w, h = 800, 500
screen = pygame.display.set_mode((w, h))

# Take image as input
img = pygame.image.load("../Python/Input/GeeksforGeeks.png")
img.convert()

# Draw rectangle around the image
rect = img.get_rect()
rect.center = w//2, h//2

# Set running and moving values
running = True
moving = False

# Setting what happens when game 
# is in running state
while running:
	
	for event in pygame.event.get():

		# Close if the user quits the 
		# game
		if event.type == QUIT:
			running = False

		# Making the image move
        # collidepoint(x, y) -> bool: test if a point is inside a rectangle
		elif event.type == MOUSEBUTTONDOWN:
			if rect.collidepoint(event.pos):
				moving = True

		# Set moving as False if you want 
		# to move the image only with the 
		# mouse click
		# Set moving as True if you want 
		# to move the image without the 
		# mouse click
		elif event.type == MOUSEBUTTONUP:
			moving = False

		# Make your image move continuously
		elif event.type == MOUSEMOTION and moving:
			rect.move_ip(event.rel)

	# Set screen color and image on screen
	screen.fill(YELLOW)
	screen.blit(img, rect)

	# Construct the border to the image
	pygame.draw.rect(screen, BLUE, rect, 2)

	# Update the GUI pygame
	pygame.display.update()

# Quit the GUI game
pygame.quit()
```

## Note
blit(src, dest)
- Blitting is the process of rendering the game object onto the surface
- src : Source of the image which we want to display on the screen
- dest : Coordinates where we want our image to be displayed.

rect(left, top, width, height)
- to store and manipulate rectangular areas

pygame.draw.rect(surface, color, rect, width)

pygame.display.update()
 - Update portions of the screen for software displays.
 - If no argument is passed it updates the entire Surface area like pygame.display.flip()
