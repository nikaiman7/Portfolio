## Produce Exchange Rates

```python
# Step 1: Get the API response
url      = "https://api.exchangeratesapi.io/v1/latest?access_key=9cceff5e4230aecf4467861c051fa654&format=1" # API endpoint
response = requests.get(url)                                                                                # Send GET request
data     = response.json()                                                                                  # Convert response JSON string to Python dict

# Step 2: Extract and sort 'rates' by ascending value
rates        = data["rates"]
sorted_rates = dict(sorted(rates.items(), key=lambda item: item[1], reverse=True))

# Code walkthrough
# rates.items()            : Converts the dictionary into a list of (key, value) tuples   [('USD', 1.15), ('JPY', 166.42), ('BTC', 0.00001)]
# sorted()                 : Sorts that list of tuples.
# key=lambda item: item[1] : Tells Python to sort based on the exchange rate value (not the key).
# dict(...)                : Converts the sorted list of tuples back into a dictionary:

# Step 3: Convert to DataFrame
df = pd.DataFrame(sorted_rates.items(), columns=["Currency", "Rate"])

# Step 4: Preview original data & DataFrame
print("Original Data Directly from API:")
print(data)

print("\nDataframe")
display(df)
```

```python
# Plot
# Select lowest and highest N currencies
n = 10
df_low  = df.head(n).copy()
df_high = df.tail(n).copy()

# Add labels for category
df_low["Category"]  = "Lowest"
df_high["Category"] = "Highest"
df_combined = pd.concat([df_low, df_high])

# Sort combined for better visual layout
df_combined = df_combined.sort_values("Rate")

# Plot with log scale
plt.figure(figsize=(10, 8))
colors = ["tomato" if cat == "Lowest" else "seagreen" for cat in df_combined["Category"]]
plt.barh(df_combined["Currency"], df_combined["Rate"], color=colors)
plt.xscale("log")
plt.title(f"Top {n} Highest and Lowest Exchange Rates (vs EUR) [Log Scale]")
plt.xlabel("Exchange Rate (Log Scale)")
plt.ylabel("Currency")
plt.tight_layout()
plt.show()
```

# Webscrapping - Banks from Wikipedia
On Wikipedia, it looks like this: <br>


In HTML, it looks like this: <br>


```python
# Step 1: Fetch the HTML content
URL       = "https://en.wikipedia.org/wiki/List_of_largest_banks"
html_data = requests.get(URL).text

# Step 2: Parse the HTML with BeautifulSoup
soup      = BeautifulSoup(html_data,'html.parser')

# Step 3: Find all tables with class 'wikitable'
tables = soup.find_all('table', {'class': 'wikitable'})

# Step 4: Convert the N table to a DataFrame using StringIO
n          = 0
table_html = str(tables[n])
dfs        = pd.read_html(StringIO(table_html))
df         = dfs[0]

# Note
# n=0                  : Table for total assets
# n=1                  : Table for number of banks in the top 100 by total assets
# n=2                  : Table for market capitalization

# str(tables[0])       : Takes the first <table> from the list and converts it into an HTML string.
# StringIO(table_html) : Wraps that HTML string into a file-like object, as expected by pd.read_html()
# pd.read_html(...)    : Scans the HTML for <table> tags and automatically converts them into pandas DataFrames.
#                        It returns a list of DataFrames. Often there's only one, so we take the first with dfs[0].

# display(soup)
display(df)
```

# ETL
Objectives:
- Run the ETL process
    - ***Extract*** bank and market cap data from the JSON file (bank_market_cap.json)
    - ***Transform*** the market cap currency using the CSV file (exchange_rates.csv)
    - ***Load*** the transformed data into a seperate CSV file (bank_market_cap_gbp.csv)

## Functions

```python
# Doesn't work in Jupyter, proposed workarounds is as follows
# !wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0221EN-SkillsNetwork/labs
# /module%206/Lab%20-%20Extract%20Transform%20Load/data/bank_market_cap_1.json
# !wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0221EN-SkillsNetwork/labs
# /module%206/Lab%20-%20Extract%20Transform%20Load/data/bank_market_cap_2.json
# !wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0221EN-SkillsNetwork/labs
# /module%206/Final%20Assignment/exchange_rates.csv

# Function #1: Extract Function
def extract_json(extract_file):
    # Define and wrap the URL for readability
    json_url = (
                "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/"
                "IBMDeveloperSkillsNetwork-PY0221EN-SkillsNetwork/labs/module%206/"
                "Lab%20-%20Extract%20Transform%20Load/data/" + extract_file
                )
    
    # Download JSON file
    json_response = requests.get(json_url)
    with open("../Python/Input/" + extract_file, "wb") as file:
        file.write(json_response.content)

    # Load JSON into DataFrame
    df_extract_json = pd.read_json("../Python/Input/" + extract_file)
    
    return df_extract_json

def extract_csv(extract_file):
    # Define and wrap the URL for readability
    csv_url = (
                "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/"
                "IBMDeveloperSkillsNetwork-PY0221EN-SkillsNetwork/labs/module%206/"
                "Final%20Assignment/" + extract_file
               )
    
    # Download JSON file
    csv_response = requests.get(csv_url)
    with open("../Python/Input/" + extract_file, "wb") as file:
        file.write(csv_response.content)

    # Load JSON into DataFrame
    df_extract_csv = pd.read_csv("../Python/Input/" + extract_file)
    
    return df_extract_csv

# Function #2: Transform Function
def transform(bank_data,exchange_data,exchange_symbol):
    # Find exchange rate of USD to GBP
    new_currency_rate = exchange_data[exchange_data["Symbol"] == exchange_symbol]["Rates"].values[0]

    # Make a copy to avoid modifying original input
    df_transform = bank_data.copy()
    
    # Convert market cap from USD to GBP and round
    df_transform['Market Cap (US$ Billion)'] = df_transform['Market Cap (US$ Billion)'] * new_currency_rate
    df_transform['Market Cap (US$ Billion)'] = df_transform['Market Cap (US$ Billion)'].round(3)

    # Rename column
    df_transform = df_transform.rename(columns={'Market Cap (US$ Billion)': 'Market Cap (GBP$ Billion)'})

    return df_transform

# Function #3: Load Function
def load(df_name):
    # Write to csv
    df_name.to_csv('../Python/Input/bank_market_cap_gbp.csv',index=False)

# Function #4: Logging function
def log(message):
    timestamp_format = '%Y-%h-%d-%H:%M:%S'            # Year-Monthname-Day-Hour-Minute-Second
    now              = datetime.now()                 # get current timestamp
    timestamp        = now.strftime(timestamp_format)
    print(timestamp + ': ' + message)
```

## End-to-End ETL Process

```python
# Extract
# Market Caps
df_cap = extract_json('bank_market_cap_1.json')

# Exchange Rates
df_rates = extract_csv("exchange_rates.csv")
df_rates.rename(columns={"Unnamed: 0": "Symbol"}, inplace=True)

# Transform
df_cap_gbp = transform(df_cap,df_rates,"GBP")

# Load
load(df_cap_gbp)

# Display timeline
log('ETL Job Started')
log('Extract Phase Started')
print("Market Cap")
display(df_cap)
print("\nExchange Rates")
display(df_rates)
log('Extract phase Ended')
print('\n')

log('Transform phase Started')
print("\nNew Transformed Market Cap")
display(df_cap_gbp)
log('Transform phase Ended')
print('\n')

log('Load Phase Started')
log('Load Phase: File Saved!')
log('Load phase Ended')
```
